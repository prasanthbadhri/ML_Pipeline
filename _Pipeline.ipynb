```python
x_train,x_test,y_train,y_test = train_test_split(df1.iloc[:,[0,1,2,3,4,5,6,7]],df1.iloc[:,[8]],test_size=0.2,random_state=1)
```

## creating Pipeline

```python
## creating pipeline for logistic regression, Decision tree, Random forest models.
# steps for pipeline:
# 1. Data preprocessing using Minmaxscalar
# 2. Reducing Dimensionality using PCA
# 3. Traning respective models
```

```python
# logistic regression Pipeline...
LR_pipeline = Pipeline([('myscalar',MinMaxScaler()),# scaling
                       ('mypca',PCA(n_components=3)),# runs a principle component analysis on data of num of components is 3. 
                       ('logistic_classifier',LogisticRegression())])
```

```python
# Decision tree pipeline...
DT_pipeline = Pipeline([('myscalar',MinMaxScaler()),
                       ('mypca',PCA(n_components=3)),
                       ('Decisiontree_classifier',DecisionTreeClassifier())])
```

```python
# Random forest pipeline...
RF_pipeline = Pipeline([('myscalar',MinMaxScaler()),
                       ('mypca',PCA(n_components=3)), 
                       ('Randomforest_classifier',RandomForestClassifier())])
```

## model training and validation.

```python
my_pipeline = [LR_pipeline,DT_pipeline,RF_pipeline]
# defining the pipelines in a list...
```

```python
# defining a variables for choosing best model...
accuracy = 0
classifier = 0
pipeline=""
```

```python
# creating dictionary of pipelines  and training models...
pipeline_dict = {0:"Logistic Regression", 1:"Decision Tree", 2:"Random forest"}
```

```python
# fit the pipelines
for mypipe in my_pipeline:
    mypipe.fit(x_train,y_train)
```

```python
# getting test accuracy for the classifiers

for i,model in enumerate(my_pipeline):
    print("{} test_accuracy: {}".format(pipeline_dict[i],model.score(x_test,y_test)))
```

```python
# choosing the best model 

for i,model in enumerate(my_pipeline):
    if model.score(x_test,y_test)>accuracy:
        accuracy = model.score(x_test,y_test)
        pipeline = model
        classifier = i
        
print("clssifier with best accuracy:{}".format(pipeline_dict[classifier]))
```
